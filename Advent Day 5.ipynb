```python
import numpy as np
rows = []
linenum = 0
with open("AC22D5Key.txt", 'r') as f:
    for line in f.readlines():
        if line[0] == " ":
            breakline = linenum
            break
        row = []
        for count, letter in enumerate(line):
            if (count-1) % 4 == 0: #or i == 1:
                row.append(letter)
        rows.append(row)  
        linenum += 1
    Nstacks = len(row)
    A = np.array(rows)
    crates = np.count_nonzero(A)
    
print(A)
A = np.transpose(A)
A = np.flip(A, 1)
A = np.pad(A, ((0, 0), (0, crates)), 'constant', constant_values=(""))
print(A)
           
```

```python
i = 0
with open("AC22D5Key.txt", 'r') as f:
    for count, line in enumerate(f):
        if count > breakline+1:
            order = str(line).split()
            N = int(order[1])
            start = int(order[3])
            end = int(order[5])
            while i<N:
                heightstartstack = np.count_nonzero(A[start-1])
                #print(heightstartstack)
                heightendstack = np.count_nonzero(A[end-1])
                A[end-1,heightendstack] = A[start-1,heightstartstack-1]
                A[start-1,heightstartstack-1] = ""
                i += 1
            #print(A)
            i = 0
    print(A)
```
